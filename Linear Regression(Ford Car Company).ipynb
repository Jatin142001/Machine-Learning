```python
X = df.drop(columns='price',axis=1)
y=df['price']
```

```python
X
```

### Showing how to do "One Hot Encoding" on model, transmission, and fueltype columns.

```python
X_one_hot_encoding = pd.get_dummies(X,columns=['model','transmission','fuelType'],drop_first=False)
```

```python
X_one_hot_encoding
```

```python
X_one_hot_encoding = X_one_hot_encoding.astype(int)
```

```python
X_one_hot_encoding
```

### Showing how to do "Label Encoding" on model, transmission, and fueltype columns.

```python
from sklearn.preprocessing import LabelEncoder

columns = ['model','transmission','fuelType']
X_label = X.copy()
label_encoders = {}

for col in columns:
    encode = LabelEncoder()
    X_label[col] = encode.fit_transform(X_label[col].astype(str))
    label_encoders[col] = encode
```

```python
X_label
```

### Converting some columns into standard scaling between -3 and +3 values.

```python
from sklearn.preprocessing import StandardScaler

numeric_cols = ['year','mileage','tax','mpg','engineSize']
scaler = StandardScaler()
X_one_hot_encoding[numeric_cols] = scaler.fit_transform(X_one_hot_encoding[numeric_cols])
```

```python
X_one_hot_encoding
```

### Converting all the columns into standard scaling for label encoding. 

```python
X_label.columns
```

```python
X_label[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']] = scaler.fit_transform(X_label[['model', 'year', 'transmission', 'mileage', 'fuelType', 'tax', 'mpg',
       'engineSize']])
```

```python
X_label
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot_encoding, y, test_size=0.20, random_state=42)
```

```python
model = LinearRegression()

model.fit(X_train,y_train)
```

```python
y_prediction = model.predict(X_test)
```

```python
y_prediction
```

```python
y_test
```

```python
r2 = r2_score(y_test,y_prediction)
r2
```

```python
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r2 = 1 - ((1-r2)*(n-1))/(n-p-1)
print("Adjusted R2 score: ",adjusted_r2)
```
